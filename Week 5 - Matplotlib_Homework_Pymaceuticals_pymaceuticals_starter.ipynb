```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "../Week 5 - Matplotlib_Homework_Pymaceuticals_data_mouse_drug_data.csv"
clinical_trial_data_to_load = "../Week 5 - Matplotlib_Homework_Pymaceuticals_data_clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset
all_data = pd.merge(trial_data, mouse_data, on="Mouse ID", how="left")


# Display the data table for preview
all_data
```

# Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_mean = all_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()


# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_mean).reset_index()

# Preview DataFrame
tumor_mean_df.head(20)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = all_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_sem).reset_index()

# Preview DataFrame
tumor_sem_df.head(20)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_mean_pivot = tumor_mean_df.pivot(index= "Timepoint", columns= "Drug", values= "Tumor Volume (mm3)")
tumor_sem_pivot = tumor_sem_df.pivot(index= "Timepoint", columns= "Drug", values= "Tumor Volume (mm3)")

# Preview that Reformatting worked
tumor_mean_pivot
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Capomulin"],yerr=tumor_sem_pivot["Capomulin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Capomulin")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Ceftamin"],yerr=tumor_sem_pivot["Ceftamin"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ceftamin")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Infubinol"],yerr=tumor_sem_pivot["Infubinol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Infubinol")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Ketapril"],yerr=tumor_sem_pivot["Ketapril"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ketapril")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Naftisol"],yerr=tumor_sem_pivot["Naftisol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Naftisol")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Placebo"],yerr=tumor_sem_pivot["Placebo"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Placebo")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Propriva"],yerr=tumor_sem_pivot["Propriva"], 
             linewidth=1, marker="o", markersize=3, linestyle='--',label="Propriva")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Ramicane"],yerr=tumor_sem_pivot["Ramicane"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Ramicane")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Stelasyn"],yerr=tumor_sem_pivot["Stelasyn"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Stelasyn")
plt.errorbar(tumor_mean_pivot.index,y=tumor_mean_pivot["Zoniferol"],yerr=tumor_sem_pivot["Zoniferol"], 
             linewidth=1,marker="o",markersize=3,linestyle='--',label="Zoniferol")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.xticks(np.arange(0, 75, 5))
plt.grid()
plt.legend(loc="best")


# Save the Figure
plt.savefig("Tumor Repsonse Over Time.png")
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = all_data.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
mouse_df = pd.DataFrame(mouse_count).reset_index()

# Preview DataFrame
mouse_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = mouse_df.pivot(index= "Timepoint", columns= "Drug", values= "Mouse ID")

# Preview that Reformatting worked
mouse_count_pivot
```

```python
# Generate the Plot (Accounting for percentages)
survival_rate = round(mouse_count_pivot / (mouse_count_pivot.iloc[0]) * 100, 2)
survival_rate

plt.plot(survival_rate.index,survival_rate["Capomulin"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Capomulin")
plt.plot(survival_rate.index,survival_rate["Ceftamin"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Ceftamin")
plt.plot(survival_rate.index,survival_rate["Infubinol"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Infubinol")
plt.plot(survival_rate.index,survival_rate["Ketapril"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Ketapril")
plt.plot(survival_rate.index,survival_rate["Naftisol"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Naftisol")
plt.plot(survival_rate.index,survival_rate["Placebo"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Placebo")
plt.plot(survival_rate.index,survival_rate["Propriva"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Propriva")
plt.plot(survival_rate.index,survival_rate["Ramicane"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Ramicane")
plt.plot(survival_rate.index,survival_rate["Stelasyn"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Stelasyn")
plt.plot(survival_rate.index,survival_rate["Zoniferol"],
         linewidth=1,marker="o",markersize=3,linestyle='--',label="Zoniferol")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.xticks(np.arange(0, 75, 5))
plt.grid()
plt.legend(loc="best")

# Save the Figure
plt.savefig("Survival Rate")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = (tumor_mean_pivot.iloc[-1]/(tumor_mean_pivot.iloc[0])-1)*100

# Display the data to confirm
percent_change 
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
# Call functions to implement the function calls

passing = percent_change < 0
x_axis = np.arange(len(percent_change))


plot = plt.bar(x_axis, percent_change, color = passing.map({True:'g',False:'r'}), alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Capomulin", "Ceftamin", "Infubinol", "Ketapril",
                            "Naftisol", "Placebo", "Propriva", "Ramicane",
                           "Stelasyn", "Zoniferol"], rotation=70)
plt.xlim(-1, len(x_axis)-0.25)
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.grid()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height < 0:
            label_position = -5
        else:
            label_position = 2
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height),
                 ha='center', va='bottom')
        
autolabel(plot)


# Save the Figure
plt.savefig("Tumor Percent Change.png")

# Show the Figure
plt.show()
```

## Observable Trends

1. Analyzing the graph: "Tumor Repsonse Over Time", it can be observed that there are only 2 medications with effective results in the reduction tumor's reduction of volume over time. These two drugs are: Capomulin with a 19% reduction and Ramicane with a 22% reduction during the 45 day trial (as seen on graph "Tumor Change Over 45 Day Trial").

2. Three out of the ten medications participating during the trial, had worse results than placebos when measuring the survival rate of the mice. This medications are: Stelasyn, Infubinol and Propriva. However, only Stelasyn showed a lower volume change in tumors as compared to placebos. This indicates that, although Stelasyn and Infubinol have a slight percentage of tumor reduction, they are very aggressive for the consumer.

3. Only Ketapril, showed an increase in the number of metastases present within the mouse. So, although its survival rate was not as low as other medications, this drug is the one with the least reduction in the volume and amount of tumors.
